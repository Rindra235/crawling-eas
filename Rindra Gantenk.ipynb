```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,4):
    browser.get(f'https://www.tiket.com/to-do/country/singapore?pageNumber={index}&utm_page=toDoSearch')
    time.sleep(10)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='body')
    
    for row in rows:
        nama = row.find(class_='product-name').text
        lokasi = row.find(class_='location').text
        rating = row.find(class_='total-rating')
        if rating:
            rating = rating.text
        impress = row.find(class_='count-rating')
        if impress:
            impress = impress.text
        harga = row.find(class_='amount')
        if harga:
            harga = harga.text
        
        df = {
            'Name': nama,
            'Location': lokasi,
            'Rating': rating,
            'Review': impress,
            'Price': harga,
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
